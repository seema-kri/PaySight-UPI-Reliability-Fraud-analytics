# Statistics

```python
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind, pointbiserialr
```

```python
df = pd.read_csv("cleaned_transactions.csv")
df.shape
```

## Test 1: Is payment type actually linked to fraud?
Chi-square test checks if type and fraud are really connected, or if it just looks that way by chance.

```python
contingency = pd.crosstab(df['payment_type'], df['is_fraud'])
chi2, p_value, dof, expected = chi2_contingency(contingency)
print("Chi-square statistic:", chi2)
print("p-value:", p_value)
```

**How to read it:** small p-value (under 0.05) means type and fraud are genuinely related, not a coincidence. Got p = 0.0 and chi-square = 22,082  strong, real relationship.

## Test 2: Are fraud transactions actually bigger, on average?
T-test — compares average amount for fraud vs. legit transactions.

```python
fraud_amounts = df[df['is_fraud']==1]['money_amount']
legit_amounts = df[df['is_fraud']==0]['money_amount']
t_stat, p_value = ttest_ind(fraud_amounts, legit_amounts, equal_var=False)
print("T-statistic:", t_stat)
print("p-value:", p_value)
print("Fraud avg amount:", fraud_amounts.mean())
print("Legit avg amount:", legit_amounts.mean())
```

 small p-value means the amount difference is real, not random noise. Fraud transactions average ₹14,67,967 vs ₹1,78,197 for legit — about 8.2x bigger. This effect size matters more to the business than the p-value alone.

## Test 3: Is the balance-mismatch flag actually connected to fraud?
Correlation test — checks how strongly `balance_doesnt_match` relates to `is_fraud`.

```python
corr, p_value = pointbiserialr(df['is_fraud'], df['balance_doesnt_match'])

print("Correlation:", corr)
print("p-value:", p_value)
```
